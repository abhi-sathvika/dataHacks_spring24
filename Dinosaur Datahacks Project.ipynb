```python
# imports
import pandas as pd
from matplotlib_inline.backend_inline import set_matplotlib_formats
import matplotlib.pyplot as plt
set_matplotlib_formats("svg")
plt.style.use('ggplot')
```

```python
# load data
dino_data = pd.read_csv('datahacks_24/data.csv')
human_data  =pd.read_csv('datahacks_24/Vegetarianism_by_country_1.csv')
```

```python
# Display Dinosaur data
dino_data.head()
```

```python
# Display Human Data
human_data.head()
```

```python
# Joining data frames
merged = dino_data.merge(human_data, left_on='lived_in', right_on='Country')
merged
```

```python
# Data Cleaning/Wrangling

# Rename Columns
new = merged.assign(human_veg_prop = merged.get('Vegetarians (% of population)'))
new = new.assign(dino_diet = merged.get('diet'))
new
```

```python
# Filter Dataframe
my_sorted = new[['lived_in', 'dino_diet', 'human_veg_prop']]
my_sorted
```

```python
# Filter only to Herbivores and Carnivores
my_sorted = my_sorted[(my_sorted.get('dino_diet') == 'carnivorous') | (my_sorted.get('dino_diet') == 'herbivorous')]
my_sorted
```

```python
# Grouping by country and getting the total number of dinos in each country
num_of_dinos_by_country = my_sorted.groupby('lived_in').count()
num_of_dinos_by_country = num_of_dinos_by_country.assign(country_count = num_of_dinos_by_country.get('dino_diet'))
num_of_dinos_by_country = num_of_dinos_by_country[['country_count']]
num_of_dinos_by_country
```

```python
# Grouping by Country and Diet Type and counting the number of dinos per each diet
dino_veg_count = my_sorted.groupby(['lived_in', 'dino_diet']).count()
dino_veg_count = dino_veg_count.assign(dino_diet_count = dino_veg_count.get('human_veg_prop'))
dino_veg_count = dino_veg_count[['dino_diet_count']]
dino_veg_count
```

```python
# Cleaning/Wrangling Dataframe isolating just Herbivore diet
dino_veg = dino_veg_count.reset_index()
dino_veg = dino_veg.set_index('lived_in')
dino_veg = dino_veg[dino_veg.get('dino_diet') == 'herbivorous']
dino_veg
```

```python
# Calculating and adding the Dino (Herbivore) proportion for each country
dino_veg_prop = dino_veg.get('dino_diet_count')/num_of_dinos_by_country.get('country_count')
num_of_dinos_by_country = num_of_dinos_by_country.assign(dino_veg_prop = dino_veg_prop)
num_of_dinos_by_country
```

```python
# Calculating and adding the Dino (Carnivore) proportion for each country
dino_nv_prop = 1 - num_of_dinos_by_country.get('dino_veg_prop')
num_of_dinos_by_country = num_of_dinos_by_country.assign(dino_nv_prop = dino_nv_prop)
num_of_dinos_by_country
```

```python
# Data cleaning/wrangling removing duplicates
unique = my_sorted.drop_duplicates(subset=['human_veg_prop'])
unique = unique.set_index('lived_in')
unique
```

```python
human_dino_prop_by_country = num_of_dinos_by_country.assign(human_veg_prop = unique.get('human_veg_prop'))
human_dino_prop_by_country
```

```python
human_dino_prop_by_country = human_dino_prop_by_country.drop('United Kingdom')
human_dino_prop_by_country
```

```python
# Function to format the human vegetarian proportion column
def adjust_prop(prop):
    new_prop = prop.strip('%')
    if '–' in new_prop:
        dash_index = new_prop.index('–')
        new_prop = new_prop[:dash_index]
    return float(new_prop) /100
```

```python
# Applying the function above to the respective column
stripped = human_dino_prop_by_country.get('human_veg_prop').apply(adjust_prop)
human_dino_prop_by_country = human_dino_prop_by_country.assign(human_veg_prop = stripped)
nv_prop = 1-human_dino_prop_by_country.get('human_veg_prop')
human_dino_prop_by_country = human_dino_prop_by_country.assign(human_nv_prop = nv_prop)
```

```python
human_dino_prop_by_country
```

```python
# Visualizations
```

```python
human_dino_prop_by_country.plot(kind='bar', y=['dino_veg_prop', 'human_veg_prop']);
plt.xlabel('Country');
plt.ylabel('Proportion');
plt.title('Difference in Vegetarianism Among Dinosaurs vs Humans');
```

```python
human_dino_prop_by_country.plot(kind='bar', y=['dino_nv_prop', 'human_nv_prop']);
plt.xlabel('Country');
plt.ylabel('Proportion');
plt.title('Difference in Non-Vegetarianism Among Dinosaurs vs Humans');
```

```python
human_dino_prop_by_country.plot(kind='bar', y=['dino_veg_prop', 'human_veg_prop'], stacked=True);
plt.xlabel('Country');
plt.ylabel('Proportion');
plt.title('Difference in Vegetarianism Among Dinosaurs vs Humans');
```

```python
human_dino_prop_by_country.plot(kind='bar', y=['human_nv_prop', 'human_nv_prop'], stacked=True);
plt.xlabel('Country');
plt.ylabel('Proportion');
plt.title('Difference in Non-Vegetarianism Among Dinosaurs vs Humans');
```

```python
dino_length_diet = dino_data[['diet', 'length']]
dino_length_diet
```

```python
dino_length_diet = dino_length_diet.dropna()
dino_length_diet
```

```python
# Functioon to format length column of Dinosaurs Data
def adjust_length(length):
    new_length = length[:-1]
    return float(new_length)
```

```python
# Data Cleaning/Wrangling diet and average length columns
length_col = dino_length_diet.get('length').apply(adjust_length)
dino_length_diet = dino_length_diet.assign(average_length = length_col)
dino_length_diet = dino_length_diet[(dino_length_diet.get('diet') == 'carnivorous') | (dino_length_diet.get('diet') == 'herbivorous')]
dino_length_diet = dino_length_diet[['diet', 'average_length']]
dino_length_diet
```

```python
# Grouping by diet and computing average for each
dino_length_diet = dino_length_diet.groupby('diet').mean()
dino_length_diet
```

```python
dino_length_diet
```

```python
# Plotting
```

```python
dino_length_diet.plot(kind='bar', y='average_length', color='#0096FF');
plt.color = 'Blue';
plt.xlabel('Diet Type');
plt.ylabel('Average Length of Dinosaurs');
plt.title('Average Length of Dinosaurs on Diet Type');
```
